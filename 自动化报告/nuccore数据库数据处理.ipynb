```python
def analysisDEFINITION(contents):
    """解析DEFINITION数据"""
    res = re.match(r"[.\s\S]*DEFINITION(.*?)\n",contents)
    datas = []
    if res:
        datas.append(res.group(1))
    return datas

def analysisACCESSION(contents):
    """解析ACCESSION数据"""
    res = re.match(r"[.\s\S]*ACCESSION(.*?)\n",contents)
    datas = []
    if res:
        datas.append(res.group(1))
    return datas

def analysisVERSION(contents):
    """解析VERSION数据"""
    res = re.match(r"[.\s\S]*VERSION(.*?)\n",contents)
    datas = []
    if res:
        datas.append(res.group(1))
    return datas

def analysisORGANISM(contents):
    """解析ORGANISM数据"""
    res = re.findall(r"ORGANISM([.\s\S]*?)REFERENCE",contents)
    return res

def analysisAUTHORS(contents):
    """解析AUTHORS数据"""
    res = re.findall(r"AUTHORS([.\s\S]*?)TITLE",contents)
    return res

def analysisTITLE(contents):
    """解析TITLE数据"""
    res = re.findall(r"TITLE([.\s\S]*?)JOURNAL",contents)
    return res

def analysisJOURNAL(contents):
    """解析JOURNAL数据"""
    # 先分割REFERENCE
    spre = re.split(r"REFERENCE|COMMENT|FEATURES",contents)
#     print(spre)
    datas_list = []
    for sp in spre:
        res = re.findall(r"JOURNAL([.\s\S]*?)PUBMED|JOURNAL([.\s\S]*?)$",sp) 
#         print(res)
        for _res in res:
            for _ in _res:
                if not _:
                    continue
                datas_list.append(_)
    return datas_list

def analysisPUBMED(contents):
    """解析PUBMED数据"""
    res = re.findall(r"PUBMED([.\s\S]*?)\n",contents)
    return res

def analysisREMARK(contents):
    """解析REMARK数据"""
    res = re.findall(r"REMARK([.\s\S]*?)REFERENCE",contents)
    return res

def analysisAssemblyMethod(contents):
    """解析AssemblyMethod数据"""
    res = re.findall(r"Assembly Method.*?::([.\s\S]*?)\n",contents)
    return res

def analysisSequencingTechnology(contents):
    """解析Sequencing Technology数据"""
    res = re.findall(r"Sequencing Technology.*?::([.\s\S]*?)\n",contents)
    return res

def analysisMolType(contents):
    """解析MolType数据"""
    res = re.findall(r"mol_type=([.\s\S]*?)\n",contents)
    return res

def analysisSeroType(contents):
    """解析serotype数据"""
    res = re.findall(r"serotype=([.\s\S]*?)\n",contents)
    return res

def analysisIsolate(contents):
    """解析isolate数据"""
    res = re.findall(r"isolate=([.\s\S]*?)\n",contents)
    return res

def analysisHost(contents):
    """解析host数据"""
    res = re.findall(r"host=([.\s\S]*?)\n",contents)
    return res

def analysisTaxon(contents):
    """解析taxon数据"""
    res = re.findall(r"taxon:([.\s\S]*?)\n",contents)
    return res

def analysisCountry(contents):
    """解析country数据"""
    res = re.findall(r"country=([.\s\S]*?)\n",contents)
    return res

def analysisCollectionDate(contents):
    """解析collection_date数据"""
    res = re.findall(r"collection_date=([.\s\S]*?)\n",contents)
    return res
```

### 重命名列名

```python
def rename(df,prefix):
    """重命名列名"""
    columns = df.columns.to_list()
    new_columns = {}
    for clo in columns:
        if len(columns)>=2:
            new_columns[clo] = prefix + str(clo)
        else:
            new_columns[clo] = prefix
    df.rename(columns=new_columns,inplace=True)
```

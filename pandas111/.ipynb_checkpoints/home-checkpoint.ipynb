```python
import os
import pandas as pd
```

```python
overview_path = os.path.join(r"C:\Users\chen\Desktop\data\overview")
phenotypes_path = os.path.join(r"C:\Users\chen\Desktop\data\phenotypes")
```

```python
home_df = pd.read_excel(r"C:\Users\chen\Desktop\data\home.xlsx")
```

```python
home_df.head()
```

```python
home_df = home_df.where(home_df.notnull(),None)
```

```python
home_df
```

```python
import numpy as np
```

```python
def get_overview(info):
    if not info or info == np.nan:
        return
    try:
        path = os.path.join(overview_path,f"" + str(info).replace("/","") + ".csv")
        overview_df = pd.read_csv(path)
        if overview_df.empty:
            return
    #     display(overview_df)
        overview_text = overview_df.iloc[0]["overview"]
    except Exception as e:
        print(e.__str__())
        return
    return overview_text
```

```python
home_df["overview-Clinical"] = home_df["Syndrome Name"].apply(lambda x:get_overview(x))
```

```python
home_df
```

```python
def get_phenotypes_name(info):
    if not info or info == np.nan:
        return
    try:
        path = os.path.join(phenotypes_path,f"" + str(info).replace("/","") + ".xlsx")
        phenotypes_df = pd.read_excel(path)
        if phenotypes_df.empty:
            return
        display(phenotypes_df)
        phenotypes_df = phenotypes_df.where(phenotypes_df.notnull(),None)
        phenotypes_Name = ""
        for i in range(len(phenotypes_df)):
            if phenotypes_df.iat[i,0]:
                phenotypes_Name = str(phenotypes_df.iat[i,0]) + "," + phenotypes_Name 
        print("phenotypes_Name",phenotypes_Name)

    except Exception as e:
        print(e.__str__())
        return
    return phenotypes_Name
```

```python
def get_phenotypes_accession(info):
    if not info or info == np.nan:
        return
    try:
        path = os.path.join(phenotypes_path,f"" + str(info).replace("/","") + ".xlsx")
        phenotypes_df = pd.read_excel(path)
        if phenotypes_df.empty:
            return
        display(phenotypes_df)
        phenotypes_df = phenotypes_df.where(phenotypes_df.notnull(),None)
        phenotypes_Accession = ""
        for i in range(len(phenotypes_df)):
            if phenotypes_df.iat[i,1]:
                phenotypes_Accession = str(phenotypes_df.iat[i,1]) + "," + phenotypes_Accession
        print("phenotypes_Accession",phenotypes_Accession)
    except Exception as e:
        print(e.__str__())
        return 
    return phenotypes_Accession
```

```python
home_df["phenotypes-Name"] = home_df["Syndrome Name"].apply(lambda x:get_phenotypes_name(x))
```

```python
home_df["phenotypes-Accession"] = home_df["Syndrome Name"].apply(lambda x:get_phenotypes_accession(x))
```

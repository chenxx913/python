```python
from docx import Document
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx import shared
from docx.shared import RGBColor #设置字体的颜色
from docx.shared import Cm, Inches,Pt
from docx.enum.table import WD_TABLE_ALIGNMENT
from docx.enum.style import WD_STYLE
from docx.enum.table import WD_ALIGN_VERTICAL
import pandas as pd
from scipy import optimize

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import MultipleLocator

# 支持中文
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

```python
def update_font(paragraph_format,is_center=True):
    """更新字体"""
#     paragraph_format=doc.paragraphs[0].paragraph_format
#     paragraph_format.line_spacing = Pt(18)
    paragraph_format.space_before = Pt(0)
    paragraph_format.space_after = Pt(0)
    if is_center:
        paragraph_format.alignment = WD_ALIGN_PARAGRAPH.CENTER
    
    
def update_run(run,font_size=Pt(12)):
    """更新字体"""
    run.font.name = u'宋体'
    run.font.color.rgb = RGBColor(0,0,0)#设置颜色为黑色
    run.font.size = font_size #设置字体大小为12磅 相当于 小四
    
```

### 数据处理

```python


def line_func(x, A, B):
    """定义一条直线方程"""
    return A * x + B

def get_curve(x_list,y_list):
    """根据标准曲线计算新的斜率"""
    # 拟合一条标准曲线
    x, k = optimize.curve_fit(line_func, x_list, y_list)[0]
    return x, k
    # 根据标准曲线计算新的斜率值
    y_re = []
    for x_item in x_list:
        y_re.append(x * x_item + k)
    return y_re

def peason_computer(x,y):
    """相关系数"""
    df = pd.DataFrame()
    df["x"] = x
    df["y"] = y
    pearson = df["x"].corr(df["y"])
    pearson = pearson * pearson
    return pearson
```

```python
def create_chart(x,y,chart_name,title,xlabel,ylabel):
    """生成一个图表"""
    df = pd.DataFrame()
    df["x"] = x
    df["y"] = y
    # 设置大小
    plt.rcParams['figure.figsize'] = (8,3.5)
    # 设置画布
    
    fig = plt.figure() # figsize=(4,4.5)
    fig.patch.set_facecolor('white')
    
    # parameter = np.polyfit(x, y, 2)
    # f = np.poly1d(parameter) # 拼接方程
    plt.scatter(x,y)
    # 预测直线
    a,k = get_curve(x,y)
    curve_df = pd.DataFrame()
    curve_df["x"] = x
    curve_df["y"] = curve_df["x"] * a + k
    plt.plot(x,curve_df["y"].to_list())
    line = "y = {:.3f}x + ({:.1f})".format(a,k)
    # 计算R方
    R2 ="R2 = {:.4f}".format(peason_computer(x,y))

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    line_R2 = line + "\n" + R2
    plt.annotate(line_R2,xy=(x[int(len(x)/2)],2))

    #把x轴的主刻度设置为1的倍数
    # y_major_locator=MultipleLocator(100)
    # ax=plt.gca()
    # ax.yaxis.set_major_locator(y_major_locator)

    plt.ylim(-200)
    plt.grid()
    plt.savefig("{}.png".format(chart_name),dpi=500,bbox_inches = 'tight',pad_inches=0)
```

```python
x = [0.3,0.26,0.23,0.2,0.16,0.13,0.1,0.06,0.03]
y = [1230.751852,1220.285316,1097.437812,865.2329169,588.6443478,520.0213975,376.016705,164.6923874,0]
title = "Taq酶活性标准曲线"
xlable = "Taq(U/ul)"
ylable = "斜率"
chart_name = "taq预测"
create_chart(x,y,chart_name,title,xlable,ylable)
```
